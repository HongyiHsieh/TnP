```python
import awkward as ak
from pathlib import Path
import numpy as np
import matplotlib.pyplot as plt
import mplhep as hep
import hist
from hist import Hist

hep.style.use("CMS")
palette = ["#5790fc", "#f89c20", "#e42536", "#964a8b", "#9c9ca1", "#7a21dd"]
```

```python
# load the data
data_arr = ak.from_parquet('./parquet/Data_mass.parquet')

# load the data
MC_arr = ak.from_parquet('./parquet/DY_mass.parquet')
```

```python
print(sum(data_arr.Data_count))

print(sum(MC_arr.DY_count))
```

```python
plots_dir = './plots'
Path(plots_dir).mkdir(exist_ok=True)

# Define the binning
n_bins = 100
x_low = 60
x_high = 120
binning = np.linspace(x_low, x_high, n_bins + 1)
width = binning[1] - binning[0]
center = (binning[:-1] + binning[1:]) / 2
```

```python
xerrs = [width * 0.5 for _ in range(n_bins)]
yerrs = np.sqrt(data_arr.Data_count)

yerrs = yerrs / len(data_arr.Data_count)

data_over_MC_weight = sum(data_arr.Data_count) / sum(MC_arr.DY_count)

mc_mass_hist = data_over_MC_weight * MC_arr.DY_count
```

```python
print(np.sqrt(data_arr.Data_count))
```

```python
plt.errorbar(center, data_arr.Data_count, xerr=xerrs, yerr=yerrs, linestyle="None", color="black", marker="o", label="Data")
plt.hist(center, bins=binning, weights=mc_mass_hist, histtype='step', label='MC')

hep.cms.label("Preliminary",loc=2,com=13)

plt.xlabel('Invariant mass of $e^+e^-$ system (GeV)')
plt.ylabel('count')
plt.legend()
plt.xlim(x_low, x_high)
plt.ylim(bottom=0)
plt.title('Mass Distribution')
plt.savefig(f'{plots_dir}/zee_mass.png')
plt.show()
```
